## **Call cost per person**

```python
infobank_df = pd.read_csv('bank-additional-full.csv', sep=';')
```

```python
yes_tel = (infobank_df['y'] == 'yes') & (infobank_df['contact'] == 'telephone')
cost_yes = 0.6*(infobank_df['duration'][yes_tel]/60.)*infobank_df['campaign'][yes_tel]

yes_cel = (infobank_df['y'] == 'yes') & (infobank_df['contact'] == 'cellular')
cost_yes = pd.concat([cost_yes, 0.8*(infobank_df['duration'][yes_cel]/60.)*infobank_df['campaign'][yes_cel]])
```

```python
Q1, Q3 = cost_yes.quantile(0.25), cost_yes.quantile(0.75)
IIQ = 1.5*(Q3 - Q1)

Q_lower = Q1 - IIQ
Q_upper = Q3 + IIQ

ii = (cost_yes > Q_lower) & (cost_yes < Q_upper)
cost_yes = cost_yes[ii]
cost_yes.describe()
```

```python
sns.distplot(cost_yes)
plt.xlabel('Custo de operação por cliente ativo (R$)')
plt.ylabel('Distribuição de probabilidade')
```

```python
no_tel = (infobank_df['y'] == 'no') & (infobank_df['contact'] == 'telephone')
cost_no = 0.6*(infobank_df['duration'][no_tel]/60.)*infobank_df['campaign'][no_tel]

no_cel = (infobank_df['y'] == 'no') & (infobank_df['contact'] == 'cellular')
cost_no = pd.concat([cost_no, 0.8*(infobank_df['duration'][no_cel]/60.)*infobank_df['campaign'][no_cel]])
```

```python
Q1, Q3 = cost_no.quantile(0.25), cost_no.quantile(0.75)
IIQ = 1.5*(Q3 - Q1)

Q_lower = Q1 - IIQ
Q_upper = Q3 + IIQ

ii = (cost_no > Q_lower) & (cost_no < Q_upper)
cost_no = cost_no[ii]
cost_no.describe()
```

```python
sns.distplot(cost_no)
plt.xlabel('Custo de operação por cliente não-ativo (R$)')
plt.ylabel('Distribuição de probabilidade')
```

```python
cost_yes_upper = cost_yes.quantile(0.9)
cost_no_upper = cost_no.quantile(0.9)
print('Custo máximo por operação positiva R$ %.2f' % cost_yes_upper)
print('Custo máximo por operação negativa R$ %.2f' % cost_no_upper)
```

```python
def profit_calc(Y_pred, Y_test, number_people=3*1e6, 
                cost_yes_upper=cost_yes_upper, cost_no_upper=cost_no_upper):
    
    ## Positive contact 
    ii = (Y_pred == 1) & (Y_test == 1)
    inpay = 120*ii.sum()
    outpay = cost_yes_upper*ii.sum()
    profit_mean = (120 - cost_yes_upper)*ii.sum()
    
    ## Negative contact
    ii = (Y_pred == 1) & (Y_test == 0)
    outpay = outpay + cost_no_upper*ii.sum()
    profit_mean = profit_mean - cost_no_upper*ii.sum()
    
    # Precision of the model
    ii = (Y_pred == 1) & (Y_test == 1)
    precision = ii.sum()/((Y_pred==1).sum())
    
    # Splitting between potential and not potential clients
    ii = Y_pred == 1
    split = ii.sum()/len(Y_pred)
    
    ## Profit per contacted person            
    ii = (Y_pred == 1)
    inpay_person = inpay/ii.sum()
    outpay_person = outpay/ii.sum()
    profit_person = profit_mean/ii.sum()
    
    
    ## Total estimated profit    
    inpay_estimated = (split*number_people*inpay_person)/1e6
    outpay_estimated = (split*number_people*outpay_person)/1e6
    profit_estimated = (split*number_people*profit_person)/1e6
    
    
    print('Precision: %.3f' % precision)
    print('Split: %.3f' % split)
    print('Profit per person: R$ %.2f\n' % profit_person)
    
    print('Total estimated inpay: R$ %.1f milhões' % inpay_estimated)
    print('Total estimated outpay: R$ %.1f milhões' % outpay_estimated)
    print('Total estimated profit: R$ %.1f milhões' % profit_estimated)
```

## **Targets and features**

```python
infobank_df = pd.read_csv('database_clean.csv')
```

```python
X_sample = infobank_df.drop(['duration', 'y', ], axis=1, inplace=False)
Y_sample = infobank_df['y']
```

```python
ii = X_sample.duplicated()
X_sample = X_sample[~ii]
Y_sample = Y_sample[~ii]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X_sample, Y_sample, test_size=0.3, stratify=Y_sample, random_state=0)
```

### **The default model**: call for everyone

```python
## Getting potential clients
Y_pred = np.ones_like(Y_test)
profit_calc(Y_pred, Y_test)
```

### **The perfect model**: call only for clients who will be accept that service

```python
## Getting potential clients
Y_pred = Y_test
profit_calc(Y_pred, Y_test)
```

```python
pip_LogR = Pipeline([
    ('ohe', OneHotEncoder(use_cat_names=True)),
    ('scaler', StandardScaler()),
    ('resample', RandomOverSampler()),
    ('estimator', LogisticRegression(penalty='l2', max_iter=200, solver='lbfgs', class_weight=None,
                                     n_jobs=-1, random_state=0))
])
```

```python
## Getting potential clients
Y_pred = pip_LogR.predict(X_test)
profit_calc(Y_pred, Y_test)
```

## **Probability function**

```python
def profit_calc_prob(Y_prob, prob_min, Y_test, number_people=3*1e6, 
                cost_yes_upper=cost_yes_upper, cost_no_upper=cost_no_upper):
    
    ## Positive contact 
    ii = (Y_prob >= prob_min) & (Y_test == 1)
    profit_mean = (120 - cost_yes_upper)*ii.sum()
    
    ## Negative contact
    ii = (Y_prob >= prob_min) & (Y_test == 0)
    profit_mean = profit_mean - cost_no_upper*ii.sum()
    
    # Precision of the model
    ii = (Y_prob >= prob_min) & (Y_test == 1)
    precision = ii.sum()/((Y_pred==1).sum())
    
    # Splitting between potential and not potential clients
    ii = Y_prob >= prob_min
    split = ii.sum()/len(Y_pred)
    
    ## Profit per contacted person            
    ii = (Y_prob >= prob_min)
    profit_person = profit_mean/ii.sum()
    
    
    ## Total estimated profit    
    profit_estimated = (split*number_people*profit_person)/1e6
    
    return profit_estimated
```

```python
Y_prob = pip_LogR.predict_proba(X_test)
prob_vec = np.linspace(0.0, 1.0, num=101)
```

```python
profit_dpf = []
profit_lowcost = []
profit_highcost = []

for i in np.arange(len(prob_vec)):
    
    profit_dpf.append(profit_calc_prob(Y_prob[:,1], prob_vec[i], Y_test))
    profit_lowcost.append(profit_calc_prob(Y_prob[:,1], prob_vec[i], Y_test,
                                           cost_yes_upper=cost_no_upper/2, cost_no_upper=cost_no_upper/2))
    profit_highcost.append(profit_calc_prob(Y_prob[:,1], prob_vec[i], Y_test,
                                            cost_yes_upper=2*cost_yes_upper,
                                            cost_no_upper=2*cost_no_upper))
    
profit_dpf = np.array(profit_dpf)
profit_lowcost = np.array(profit_lowcost)
profit_highcost = np.array(profit_highcost)
```

```python
plt.figure(figsize=(8,5))
plt.plot(prob_vec, profit_dpf)
plt.plot(prob_vec, profit_lowcost)
plt.plot(prob_vec, profit_highcost)
plt.legend(['Depósito a prazo fixo', 'Baixo custo operacional', 'Alto custo operacional'], 
           loc='lower right')
plt.xlabel('Probabilidade mínima de conversão')
plt.ylabel('Lucro líquido (milhões de R$)')
```
