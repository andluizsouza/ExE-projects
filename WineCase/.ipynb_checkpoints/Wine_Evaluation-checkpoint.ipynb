# **WINE CASE: Evaluation**

```python
import sqlite3
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
wine_db = sqlite3.connect('wine.db')
cursor = wine_db.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
ficha_df = pd.read_sql_query("SELECT * from ficha_tecnica", wine_db)
```

```python
pot_anos = []

for pot_string in ficha_df['potencial_de_guarda']:
    if ('ano' not in pot_string):
        print('ERROR: potencial_de_guarda nao esta em anos:')
        print(pot_string)
    else:
        pot_anos.append(pot_string.strip('anos '))
        
pot_anos = np.array(pot_anos, 'int')
ficha_df['potencial_anos'] = pot_anos
```

## **Baseline Model**

```python
potencial_medio = ficha_df.groupby(['tipo', 'pais'])['potencial_anos'].mean()
```

```python
X_sample = pd.concat([ficha_df['tipo'], ficha_df['pais']], axis=1)
```

```python
pot_range = [0,5,10,20,50]
Y_sample = np.zeros(len(ficha_df))

for i in range(len(pot_range)-1):
    ii = (ficha_df['potencial_anos'] > pot_range[i]) & (ficha_df['potencial_anos'] <= pot_range[i+1])
    Y_sample[ii] = i
```

```python
def baseline_predict(x_test):
    
    y_pred = []

    for i in x_test.index:
        pot_pred = potencial_medio[x_test['tipo'][i]][x_test['pais'][i]]
        
        for j in range(len(pot_range)-1):
            if (pot_pred > pot_range[j]) & (pot_pred <= pot_range[j+1]):
                pot_bin = j
                break
                
        y_pred.append(pot_bin)
            
    y_pred = np.array(y_pred)        
    return y_pred
```

```python
Y_pred = baseline_predict(X_sample)
score_baseline = (Y_sample == Y_pred).sum()/len(Y_sample)
print('Baseline model score:', score_baseline.round(4))
```

```python
price_df = pd.read_sql_query("SELECT * from price", wine_db)
```

```python
ii = Y_pred != Y_sample
prejuizo_nonpartner_base = price_df['price_non_partner'][ii].sum()
print("Prejuizo por erro de previsão do modelo baseline")
print("Nao parceiros: R$", prejuizo_nonpartner_base.round(2))
prejuizo_partner_base = price_df['price_partner'][ii].sum()
print("Parceiros:     R$", prejuizo_partner_base.round(2))
```
